## 채점 & 진단 결과

14문제를 모두 입력했다면 아래 셀을 실행하세요.

```python
answer_key = {1: 2, 2: 3, 3: 3, 4: 2, 5: 1, 6: 2, 7: 2, 8: 3, 9: 2, 10: 1, 11: 2, 12: 2, 13: 2, 14: 3}
area_by_q = {1: 'EDA', 2: 'EDA', 3: '전처리', 4: '전처리', 5: '시각화', 6: '시각화', 7: '회귀', 8: '회귀', 9: '분류', 10: '분류', 11: '심화', 12: '심화', 13: '딥러닝', 14: '딥러닝'}
area_guide = {'EDA': '02_데이터로딩_EDA 폴더 + 프리미엄 가이드북 8장', '전처리': '04_데이터전처리 폴더 + 프리미엄 가이드북 9장', '시각화': '03_데이터시각화 폴더 + 프리미엄 가이드북 10장', '회귀': '05_머신러닝_회귀 폴더 + 프리미엄 가이드북 11장', '분류': '06_머신러닝_분류 폴더 + 프리미엄 가이드북 12장', '심화': '07_머신러닝_심화 폴더 + 프리미엄 가이드북 13장', '딥러닝': '08_딥러닝 폴더 + 프리미엄 가이드북 14장'}

my_answers = {
    1: my_answer_1,
    2: my_answer_2,
    3: my_answer_3,
    4: my_answer_4,
    5: my_answer_5,
    6: my_answer_6,
    7: my_answer_7,
    8: my_answer_8,
    9: my_answer_9,
    10: my_answer_10,
    11: my_answer_11,
    12: my_answer_12,
    13: my_answer_13,
    14: my_answer_14
}

score = 0
wrong_areas = {}
for q, my_ans in my_answers.items():
    correct = answer_key[q]
    area = area_by_q[q]
    if my_ans == correct:
        score += 1
    else:
        wrong_areas.setdefault(area, 0)
        wrong_areas[area] += 1

print(f"=== 진단 결과: {score} / 14 문항 정답 ===")
print(f"정답률: {score/14*100:.0f}%\n")

if not wrong_areas:
    print("모든 영역에서 기본기가 탄탄합니다! 09_모의고사로 바로 실전 감각을 다지세요.")
else:
    print("아래 영역을 우선순위로 복습하는 것을 추천합니다 (틀린 문항이 많은 순):")
    for area, cnt in sorted(wrong_areas.items(), key=lambda x: -x[1]):
        print(f"  - [{area}] {cnt}문항 오답 -> {area_guide[area]}")
```

## 진단 결과 활용법

- **정답률 80% 이상**: 기본기가 충분합니다. PART 1~2 가이드북만 가볍게 훑고 09_모의고사부터 시작해도 좋습니다.
- **정답률 50~80%**: 약한 영역 1~2개만 집중 복습한 뒤 21장 14일 로드맵을 따라가세요.
