시험 공부를 하다 보면 내가 짠 코드가 맞는지 헷갈릴 때가 많습니다. 그럴 때 `aice_grader`를 부르면 인공지능 선생님처럼 즉시 정답을 알려줍니다.

아래의 셀들을 위에서부터 차례대로 실행(`Shift + Enter`)해 보세요!

```python
import pandas as pd
import numpy as np

# 1. 먼저 채점기 모듈을 불러옵니다!
import aice_grader as grader
```

### ❌ 틀렸을 때 어떻게 나올까?
일부러 결측치(NaN)가 들어있는 불량 데이터를 만들어 보겠습니다.

```python
# 결측치가 있는 데이터 준비
df = pd.DataFrame({'이름': ['철수', '영희', '민수'], '나이': [20, np.nan, 22]})
display(df)

# 채점기야, 결측치 처리가 완벽히 끝났는지 확인해줘!
grader.check_nulls(df)
```

### ✅ 맞았을 때 어떻게 나올까?
이제 결측치를 평균 나이로 꽉 채워(fillna) 전처리를 완벽하게 끝낸 후 다시 채점해 보겠습니다.

```python
# 결측치를 평균 나이(21)로 채우기
df['나이'] = df['나이'].fillna(21)
display(df)

# 다시 한번 채점기 실행!
grader.check_nulls(df)
```
