## 3교시: 데이터 전처리

**문제 5.** `target`을 y로, 나머지를 X로 하여 `train_test_split(test_size=0.2, stratify=y, random_state=42)`로 분할하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from sklearn.model_selection import train_test_split
X = digits.drop(columns=['target'])
y = digits['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)  # 데이터를 모델 학습용(train)과 검증용(test)으로 분리
print(X_train.shape)
```

</details>

**문제 6.** `StandardScaler`를 X_train/X_test에 적용하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()  # 데이터를 평균 0, 표준편차 1이 되도록 스케일링 (거리 기반, 딥러닝 알고리즘에 필수)
X_train_s = scaler.fit_transform(X_train)  # 훈련 데이터로 스케일링 기준을 학습(fit)하고 변환(transform)까지 수행
X_test_s = scaler.transform(X_test)  # 훈련 데이터에서 학습된 기준을 그대로 적용하여 평가 데이터를 변환 (데이터 누수 방지)
```

</details>

## 4교시: 머신러닝 모델링

**문제 7.** `RandomForestClassifier(n_estimators=200, random_state=42)`를 학습시키고 accuracy와 macro f1을 구하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score
rf = RandomForestClassifier(n_estimators=200, random_state=42)  # 여러 개의 의사결정 나무를 묶어 예측하는 랜덤 포레스트 분류 모델 생성
rf.fit(X_train_s, y_train)
pred = rf.predict(X_test_s)
print(accuracy_score(y_test, pred))  # 전체 예측 중 올바르게 맞춘 비율(정확도) 계산
print(f1_score(y_test, pred, average='macro'))  # 정밀도와 재현율의 조화 평균, 데이터가 불균형할 때 중요한 평가지표
```

</details>

## 5교시: 딥러닝 모델링

**문제 9.** 출력층 노드=10(softmax)인 Sequential 모델을 만들고 `sparse_categorical_crossentropy`로 compile하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()  # 딥러닝 층(Layer)을 순서대로 쌓기 위한 기본 틀(모델) 생성
model.add(Dense(64, activation='relu', input_shape=(X_train_s.shape[1],)))  # 이전 층의 모든 노드와 연결되는 완전 연결층(Dense Layer) 추가
model.add(Dense(32, activation='relu'))  # 이전 층의 모든 노드와 연결되는 완전 연결층(Dense Layer) 추가
model.add(Dense(10, activation='softmax'))  # 클래스 수(10)만큼 출력 노드를 만들고 softmax로 확률 분포 산출
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])  # 다중분류이므로 sparse_categorical_crossentropy 사용
```

</details>

**문제 10.** `EarlyStopping(patience=10)`을 사용해 학습시키고 argmax로 accuracy를 구하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import accuracy_score
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)  # 성능이 개선되지 않으면 무의미한 학습을 일찍 멈춤
model.fit(X_train_s, y_train, epochs=100, batch_size=32, validation_data=(X_test_s, y_test), callbacks=[es], verbose=0)
pred_label = np.argmax(model.predict(X_test_s), axis=1)  # 다중 분류 결과에서 각 샘플 중 가장 확률이 높은 클래스 인덱스를 추출
print(accuracy_score(y_test, pred_label))  # 전체 예측 중 올바르게 맞춘 비율(정확도) 계산
```

</details>
