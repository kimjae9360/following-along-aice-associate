## 2. Tabular 데이터 로딩

### 📖 이론 설명
실무에서는 데이터가 csv뿐 아니라 json, excel, tsv 등 다양한 형태로 제공됩니다. 각 포맷에 맞는 `read_*` 함수를 알아두면 어떤 파일이 나와도 당황하지 않습니다. 특히 한국 데이터는 엑셀에서 export한 CSV가 **cp949(euc-kr 계열)**로 인코딩된 경우가 많아, 기본값인 `utf-8`로 읽으면 한글이 깨지거나 `UnicodeDecodeError`가 납니다. 이럴 땐 `encoding='cp949'`를 시도해보는 것이 첫 번째 대처법입니다. 또, 필요한 컬럼만 미리 골라 불러오고 싶다면 `usecols`, 특정 컬럼을 인덱스로 바로 지정하고 싶다면 `index_col`을 씁니다.

실제 시험/실무 데이터는 CSV 한 장으로 끝나지 않는 경우도 많습니다. 엑셀 파일 하나에 시트가 여러 개 있다면 `sheet_name=`으로 원하는 시트를 골라 읽어야 하고(`sheet_name=None`이면 모든 시트를 `{시트이름: DataFrame}` 딕셔너리로 한 번에 받습니다), JSON이 `{'address': {'city': ...}}`처럼 값 안에 또 딕셔너리가 들어있는(중첩된) 구조라면 `pd.json_normalize()`로 평탄화해야 표 형태로 다룰 수 있습니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| pd.read_csv(path, encoding=, usecols=, index_col=) | CSV 로딩 |
| pd.read_json(path) | JSON 로딩 |
| pd.read_excel(path) | Excel 로딩 (openpyxl 필요) |
| pd.read_table(path) | TSV/공백구분 로딩 |
| pd.read_excel(path, sheet_name='시트명') | 여러 시트 중 하나만 지정해서 로딩 |
| pd.read_excel(path, sheet_name=None) | 모든 시트를 {시트명: DataFrame} 딕셔너리로 로딩 |
| pd.json_normalize(data) | 중첩된 JSON(딕셔너리 안의 딕셔너리)을 평탄한 표로 변환 |

### 💻 예제 코드

````
# 한글이 깨진다면 encoding='cp949'를 먼저 시도
df_train = pd.read_csv('../data/train.csv', encoding='utf-8')

# 필요한 컬럼만 골라서, 인덱스까지 지정해서 불러오기
passenger_df = pd.read_csv('../data/train.csv',
                            usecols=['PassengerId', 'Name', 'Sex', 'Age'],
                            index_col='PassengerId')
passenger_df.head()

df_json = pd.read_json('../data/data.json')
df_excel = pd.read_excel('../data/sample.xlsx')

# 시트가 여러 개인 엑셀: 하나만 지정해서 읽거나, None으로 전부 한번에 읽기
shops_df = pd.read_excel('../data/multi_sheet_sample.xlsx', sheet_name='매장정보')
all_sheets = pd.read_excel('../data/multi_sheet_sample.xlsx', sheet_name=None)
print(list(all_sheets.keys()))

# 중첩 JSON: address 안의 city/zipcode가 따로 컬럼으로 풀림(json_normalize)
import json
with open('../data/nested_customers.json', encoding='utf-8') as f:
    nested_data = json.load(f)
df_nested = pd.json_normalize(nested_data)
print(df_nested.columns.tolist())
````

## 3. 데이터 병합

### 📖 이론 설명
여러 개의 파일을 하나로 합쳐서 분석해야 하는 경우가 많습니다. `merge`는 공통 키를 기준으로 조인하고, `concat`은 단순히 이어붙입니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| pd.merge(a, b, on=, how=) | 키 기준 조인 |
| pd.concat([a, b]) | 이어붙이기 |

### 💻 예제 코드

````
df_test = pd.read_csv('../data/test.csv')  # PassengerId, LuggageCount, HasPet
df_all = pd.merge(df_train, df_test, on='PassengerId', how='inner')
print(df_all.shape)
df_all.head()
````

## 4. 데이터 기본 탐색

### 📖 이론 설명
본격적인 전처리에 들어가기 전에는 항상 데이터의 '뼈대'를 먼저 확인합니다. `info()`로 컬럼별 타입과 결측치 유무를, `describe()`로 수치형 통계를, `select_dtypes()`로 타입별 컬럼만 골라내는 방법을 익혀두면 다음 단계인 전처리(04번 노트북)로 바로 넘어갈 준비가 됩니다. `pd.crosstab`은 두 범주형 컬럼의 관계를 빠르게 보고 싶을 때 유용합니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| df.info() | 타입, 결측치, 메모리 |
| df.describe() / describe(include='all') | 수치형 / 전체 통계 |
| df.select_dtypes(include='number'/'object') | 타입별 컬럼 추출 |
| pd.crosstab(a, b) | 두 범주형 변수의 교차표 |

### 💻 예제 코드

````
df_train.info()
print(df_train.describe())

object_cols = df_train.select_dtypes(include='object').columns.tolist()
print('문자형 컬럼:', object_cols)

print(pd.crosstab(df_train['Sex'], df_train['Survived']))
````
