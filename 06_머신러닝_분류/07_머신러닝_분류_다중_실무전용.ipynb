## 목차

2. 평가지표 연습 (average 옵션)

````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

digits = pd.read_csv('../data/digits.csv')
X = digits.drop(columns=['target'])
y = digits['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
scaler = StandardScaler()
X_train_s = scaler.fit_transform(X_train)
X_test_s = scaler.transform(X_test)
X_train.shape
````

**문제 2.** `RandomForestClassifier(n_estimators=200, random_state=42)`를 학습시켜 예측값 `rf_pred`를 만드세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, random_state=42)  # 다중분류에서도 이진분류와 동일한 방식으로 학습(내부적으로 클래스 개수를 자동 인식)
rf.fit(X_train_s, y_train)
rf_pred = rf.predict(X_test_s)
```

</details>

## 2. 평가지표 연습 (average 옵션)

**문제 3.** `rf_pred`의 f1-score를 `average='macro'`와 `'weighted'`로 각각 구하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
from sklearn.metrics import f1_score
print(f1_score(y_test, rf_pred, average='macro'))  # macro: 클래스 크기 무시하고 단순 평균
print(f1_score(y_test, rf_pred, average='weighted'))  # weighted: 클래스별 데이터 개수로 가중 평균(다수 클래스 영향이 큼)
```

</details>
