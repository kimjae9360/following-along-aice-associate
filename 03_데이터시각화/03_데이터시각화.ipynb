## 1. 시각화 해석의 기본 원칙

### 📖 이론 설명
그래프를 그리는 것보다 **그래프를 해석하는 것**이 시험에서 더 중요합니다. 세 가지만 명심하면 대부분의 해석 문제를 풀 수 있습니다.

1. **상관관계 ≠ 인과관계.** heatmap에서 두 변수의 상관계수가 높게 나와도, 한 변수가 다른 변수의 '원인'이라고 단정할 수 없습니다. 단지 '같이 움직이는 경향이 있다'는 뜻입니다.
2. **정규분포에 가깝다 = 평균을 중심으로 좌우 대칭인 부드러운 종 모양.** 한쪽으로 치우쳐 있거나(skewed), 봉우리가 여러 개면 정규분포가 아닙니다.
3. **축이 무엇을 의미하는지 항상 먼저 확인.** 같은 모양의 그래프라도 X축/Y축이 바뀌면 해석이 완전히 달라집니다.

`plt.title()`, `plt.xlabel()`, `plt.ylabel()`은 그래프가 무엇을 보여주는지 설명하는 필수 요소이며, 시험에서도 이 옵션들을 채워야 하는 문제가 자주 나옵니다.

## 2. countplot - 범주형 변수 빈도

### 📖 이론 설명
`sns.countplot`은 범주형 컬럼의 값별 **개수(빈도)**를 막대그래프로 보여줍니다. '범주(글자)별로 값을 비교하라'는 문제가 나오면 가장 먼저 떠올려야 하는 함수입니다. `hue=` 옵션을 주면 다른 범주형 컬럼으로 막대를 한번 더 나눠서 비교할 수 있습니다(예: 성별별 생존자 수).

### 🧩 핵심 개념 정리
| 옵션 | 설명 |
|---|---|
| x= | 빈도를 셀 범주형 컬럼 |
| data= | 대상 데이터프레임 |
| hue= | 추가로 나눠 볼 범주형 컬럼 |

### 💻 예제 코드

````
sns.countplot(x='Sex', data=train)
plt.title('Count of Passengers by Sex')
plt.show()

sns.countplot(x='Sex', hue='Survived', data=train)
plt.title('Survived count by Sex')
plt.show()
````

## 3. histplot / kdeplot - 분포 보기

### 📖 이론 설명
`histplot`은 데이터가 어느 구간에 많이/적게 몰려 있는지 계단 모양(히스토그램)으로 보여줍니다. `kde=True`를 주면 그 위에 부드러운 밀도 곡선을 함께 그려 분포의 전체적인 모양(정규성, 치우침, 이상치)을 더 쉽게 볼 수 있습니다. `kdeplot`은 히스토그램 없이 곡선만 그리며, `hue=`로 그룹별 분포를 겹쳐 비교할 때 유용합니다. 출제 포인트는 '이 분포가 정규분포에 가까운가', '이상치로 보이는 값이 있는가'입니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| sns.histplot(data=, x=, kde=, bins=) | 히스토그램(+밀도곡선) |
| sns.kdeplot(data=, x=, hue=) | 밀도곡선만, 그룹 비교에 유리 |

### 💻 예제 코드

````
sns.histplot(data=train, x='Age', kde=True)
plt.title('Distribution of Age')
plt.show()

sns.kdeplot(data=train, x='Fare', hue='Survived')
plt.title('Fare distribution by Survived')
plt.show()
````

## 4. scatterplot / jointplot - 두 변수 간 관계

### 📖 이론 설명
두 연속형 변수 사이에 어떤 관계가 있는지 보고 싶을 때는 산점도(scatter)를 그립니다. `sns.jointplot`은 산점도와 함께 각 축의 히스토그램을 동시에 보여주기 때문에, 관계와 개별 분포를 한 번에 파악할 수 있습니다. `kind='reg'`를 주면 회귀선을 함께 그려 대략적인 선형 관계 유무를 보여줍니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| sns.scatterplot(data=, x=, y=) | 산점도 |
| sns.jointplot(data=, x=, y=, kind=) | 산점도+히스토그램, kind='scatter'/'reg'/'hex' |

### 💻 예제 코드

````
sns.jointplot(data=train, x='Age', y='Fare')
plt.show()

correlation = train['Age'].corr(train['Fare'])
print(f'Age-Fare 상관계수: {correlation:.4f}')  # 0에 가까우면 선형관계가 약하다는 뜻
````

## 5. heatmap - 상관관계 행렬

### 📖 이론 설명
`df.corr()`로 계산한 상관계수 표를 색상으로 시각화한 것이 heatmap입니다. `annot=True`를 빼먹으면 색만 보이고 숫자가 안 보이므로, 정확한 값을 요구하는 문제에서는 반드시 켜야 합니다. 색이 진하다고 해서 그 변수가 '원인'이라는 뜻은 아니라는 점(상관관계≠인과관계)을 다시 기억하세요.

### 🧩 핵심 개념 정리
| 옵션 | 설명 |
|---|---|
| df.corr(numeric_only=True) | 수치형 변수 간 상관계수 계산 |
| annot=True | 칸 안에 숫자 표시(필수 옵션) |
| cmap= | 색상 팔레트 |

### 💻 예제 코드

````
numeric_corr = train.select_dtypes(include=np.number).corr()
sns.heatmap(numeric_corr, annot=True, cmap='coolwarm')
plt.title('Correlation between Numeric Variables')
plt.show()
````

## 7. df.plot() / pie / 여러 그래프 나란히 그리기

### 📖 이론 설명
seaborn 없이 pandas에 내장된 `.plot()`으로도 빠르게 그릴 수 있습니다. 여러 그래프를 한 화면에 비교하고 싶을 때는 `plt.subplots(행, 열)`로 여러 개의 축(Axes)을 만들고, 각 축에 그래프를 하나씩 그립니다.

### 🧩 핵심 개념 정리
| 문법 | 설명 |
|---|---|
| series.value_counts().plot(kind='bar'/'pie') | 빠른 막대/파이차트 |
| f, ax = plt.subplots(1, 2, figsize=(12,6)) | 1행 2열 배치 |
| ax[i]에 그래프를 그리면 해당 위치에 그려짐 | |

### 💻 예제 코드

````
f, ax = plt.subplots(1, 2, figsize=(12, 6))
train['Survived'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0])
ax[0].set_title('Survived (pie)')
sns.countplot(x='Survived', data=train, ax=ax[1])
ax[1].set_title('Survived (count)')
plt.show()
````

## 9. 버블 차트 (점 크기로 세 번째 변수 표현)

### 📖 이론 설명
산점도는 두 변수의 관계만 보여주지만, 점의 **크기(size)**에 또 다른 수치형 변수를 매핑하면 **세 번째 변수까지 한 그래프에 담을 수 있습니다.** 이런 그래프를 흔히 '버블 차트'라고 부릅니다. `plt.scatter`의 `s=` 옵션에 값(보통 보기 좋게 배율을 곱한 값)을 넣고, `c=`로 색까지 다르게 하면 네 번째 변수(범주)도 함께 표현할 수 있습니다.

### 🧩 핵심 개념 정리
| 옵션 | 설명 |
|---|---|
| s= | 점 크기 - 세 번째 수치형 변수 매핑 |
| c= | 점 색상 - 네 번째 변수(범주) 매핑 |
| cmap= | 색상 팔레트 지정 |

### 💻 예제 코드
