## 6. 종합 연습 (객관식형)

**문제 8.** 아래 heatmap을 보고 `target`과 가장 상관관계가 강한(절댓값 기준) 변수를 `top_var`에 저장하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
corr = wine.corr()['target'].drop('target')  # target 자기 자신과의 상관계수(항상 1)는 제외
top_var = corr.abs().idxmax()  # abs()로 절댓값을 취해야 음의 상관관계도 '강한 관계'로 비교 가능
print(top_var)
```

</details>

## 7. 버블 차트 (점 크기로 세 번째 변수 표현)

산점도의 점 크기(`s=`)에 또 다른 수치형 변수를 매핑하면, 두 변수의 관계뿐 아니라 **세 번째 변수까지 한 그래프에 담을 수 있습니다.** (pandas_exercises의 'Scores' 문제에서 자주 나오는 방식)

**문제 1.** `iris`의 `sepal length (cm)`(x), `petal length (cm)`(y)를 산점도로 그리되, 점 크기를 `petal width (cm)`에 비례하게(`s=iris['petal width (cm)']*100`), 색은 `target`으로 다르게 하세요.

````
# 여기에 코드를 작성하세요
````

<details>
<summary>✅ 정답 보기</summary>

```python
plt.scatter(iris['sepal length (cm)'], iris['petal length (cm)'],
            s=iris['petal width (cm)'] * 100, c=iris['target'], cmap='viridis', alpha=0.6)  # s에 배율(*100)을 곱해 점 크기 차이가 잘 보이도록 조정
plt.xlabel('sepal length (cm)'); plt.ylabel('petal length (cm)')
plt.title('Bubble chart: size=petal width, color=target')
plt.show()
```

</details>
