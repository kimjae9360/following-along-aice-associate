## 1. NumPy 배열 기초

### 📖 이론 설명
NumPy 배열(ndarray)은 파이썬 리스트와 달리 **모든 원소가 같은 타입**이어야 합니다. 이 덕분에 내부적으로 C 수준의 반복문을 사용해 훨씬 빠르게 계산할 수 있습니다. 반대로 말하면, 배열 안에 문자열이 하나라도 섞이면 **전체 원소가 문자열로 강제 변환**됩니다. 시험에서 '왜 계산이 안 되지?'라는 상황의 절반은 이 dtype 문제입니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| np.array(list) | 배열 생성 |
| .dtype | 원소들의 데이터 타입 확인 |
| .shape | 배열의 형태(차원별 크기) |
| .ndim | 차원 수 |

### 💻 예제 코드

````
scores = np.array([98.0, 92.5, 84.5, 77.0, 75.5])
print(scores, scores.dtype)

# 문자열 하나가 섞이면 전체가 문자열(dtype='<U...')로 바뀐다
scores_mixed = np.array([98.0, 92.5, 84.5, 77.0, '75.5'])
print(scores_mixed, scores_mixed.dtype)
````

## 2. 배열 연산과 통계

### 📖 이론 설명
NumPy 배열은 반복문 없이 배열 전체에 연산을 적용하는 **브로드캐스팅**을 지원합니다. `arr * 2`처럼 배열과 스칼라를 바로 곱할 수 있고, `axis` 옵션으로 '행 방향'과 '열 방향' 중 어느 쪽으로 통계를 낼지 지정할 수 있습니다. `axis=0`은 '각 열(컬럼)을 따라 내려가며 계산'(→ 결과가 행 방향으로 나열), `axis=1`은 '각 행을 따라 가로질러 계산'입니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| arr.mean()/std()/sum() | 평균/표준편차/합계 |
| axis=0 | 열 기준(세로 방향) 계산 |
| axis=1 | 행 기준(가로 방향) 계산 |

### 💻 예제 코드

````
mat = np.array([[1, 2, 3], [4, 5, 6]])
print(mat.sum())          # 전체 합
print(mat.sum(axis=0))    # 열 방향 합 -> [5 7 9]
print(mat.sum(axis=1))    # 행 방향 합 -> [6 15]
print(mat * 10)           # 브로드캐스팅
````

## 3. 배열 형태 다루기

### 📖 이론 설명
`reshape`은 배열의 원소 개수는 그대로 두고 '몇 행 몇 열로 볼 것인가'만 바꿉니다. 딥러닝 파트에서 이미지를 1차원으로 펴거나(`.flatten()`), 입력 형태를 맞출 때(`input_shape`) 자주 쓰입니다. `np.random.seed()`(또는 `np.random.default_rng(seed)`)로 난수 시드를 고정하면, 코드를 몇 번 실행해도 항상 같은 결과가 나와 채점/재현이 가능해집니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| arr.reshape(r, c) | r행 c열로 형태 변경 |
| arr[start:end] | 슬라이싱 |
| np.random.seed(n) | 난수 시드 고정 |

### 💻 예제 코드

````
arr = np.arange(6)          # [0 1 2 3 4 5]
print(arr.reshape(2, 3))

np.random.seed(42)
print(np.random.randint(0, 10, size=5))  # 시드를 고정하면 항상 같은 값
````

## 5. DataFrame/Series 생성·수정·저장

### 📖 이론 설명
DataFrame은 dict로도, Series를 조합해서도 만들 수 있습니다. 여기서 자주 놓치는 함정이 하나 있습니다: **딕셔너리 값으로 Series를 넣을 때, DataFrame의 index와 Series의 index가 다르면 값이 매칭되지 않고 NaN이 됩니다.** pandas는 위치(순서)가 아니라 **인덱스 라벨**을 기준으로 값을 맞추기 때문입니다. 컬럼을 추가할 때는 `df['col']=값`으로 맨 끝에 붙거나, `insert(위치, 이름, 값)`으로 원하는 위치에 넣을 수 있습니다. `drop`은 `axis=1`(열) 또는 기본값 `axis=0`(행)을 지정하고, `inplace=True`를 주지 않으면 원본은 그대로 두고 삭제된 '새 결과'만 반환합니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| pd.DataFrame(dict, index=...) | 딕셔너리로 생성 |
| pd.Series(list, index=...) | 시리즈 생성 |
| df.insert(pos, name, value) | 원하는 위치에 컬럼 추가 |
| df.rename(columns={...}) | 컬럼명 변경 |
| df.drop(label, axis=, inplace=) | 행/열 삭제 |
| df.to_csv(path, index=False) | CSV로 저장 |

### 💻 예제 코드

````
fruit_s = pd.Series(['사과', '딸기', '참외'], index=['A', 'B', 'C'])

# index가 같으면 정상적으로 매칭됨
fruits_df = pd.DataFrame({'fruit': fruit_s, 'price': [1000, 300, 700]}, index=['A', 'B', 'C'])
print(fruits_df)

# index가 다르면 매칭이 안 되어 fruit 컬럼이 전부 NaN이 됨
fruits_df_mismatch = pd.DataFrame({'fruit': fruit_s, 'price': [1000, 300, 700]}, index=['X', 'Y', 'Z'])
print(fruits_df_mismatch)
````

## 6. 필터링과 정렬 (Filtering & Sorting)

### 📖 이론 설명
`loc`는 라벨(이름) 기준, `iloc`는 위치(정수) 기준으로 데이터를 선택합니다. 조건을 여러 개 결합할 때는 파이썬의 `and`/`or`가 아니라 `&`/`|`를 쓰고, 각 조건을 반드시 괄호로 감싸야 합니다(연산자 우선순위 때문).

### 🧩 핵심 개념 정리
| 문법 | 설명 |
|---|---|
| df.loc[조건] | 라벨/조건 기준 선택 |
| df.iloc[정수] | 위치 기준 선택 |
| (조건1) \& (조건2) | AND 조건 결합 |
| df.sort_values(by=, ascending=) | 값 기준 정렬 |

### 💻 예제 코드

````
# 30세 이상, 1등실 승객만
cond = (train['Age'] >= 30) & (train['Pclass'] == 1)
train.loc[cond, ['Name', 'Age', 'Pclass']].head()
````

## 7. 그룹화 (Grouping)

### 📖 이론 설명
`groupby`는 '기준 컬럼의 값이 같은 행들끼리 묶어서' 통계를 냅니다. 여러 컬럼으로 묶으면 결과 인덱스가 **MultiIndex**가 되는데, 특정 그룹만 보고 싶으면 `.loc[값]` 또는 `.loc[(값1, 값2)]`로 접근합니다. 집계 함수를 하나가 아니라 여러 개 동시에 적용하고 싶다면 `.agg(['mean', 'max', 'min'])`처럼 리스트로 넘기면 됩니다. `pd.crosstab`은 두 범주형 변수의 조합별 빈도를 표로 바로 보여주는, groupby보다 더 간단한 지름길입니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| df.groupby('col') | 그룹 객체 생성 |
| .groups | 그룹별 인덱스 확인 |
| .agg(['mean','max']) | 여러 집계함수 동시 적용 |
| pd.crosstab(a, b) | 두 범주형 변수의 교차표 |

### 💻 예제 코드

````
# Pclass, Embarked로 그룹화한 뒤 Fare의 평균/최대/최소를 한 번에
grouped = train.groupby(['Pclass', 'Embarked'])['Fare'].agg(['mean', 'max', 'min'])
print(grouped.loc[1])          # Pclass==1인 그룹만
print(grouped.loc[(1, 'C')])   # Pclass==1, Embarked=='C'인 그룹

# 성별 x 생존여부 교차표
print(pd.crosstab(train['Sex'], train['Survived']))
````

## 8. Apply 함수 활용

### 📖 이론 설명
`apply`는 함수를 행/열 단위로, `map`(Series 전용)은 원소 단위로 적용합니다. 실무에서 아주 자주 쓰이는 패턴은 문자열 컬럼에서 `split()`/`strip()`으로 원하는 조각만 뽑아 **새로운 파생 컬럼**을 만드는 것입니다. 예를 들어 Titanic의 `Name` 컬럼(`'Braund, Mr. Owen Harris'`)에서 호칭(Mr, Mrs, Miss 등)만 뽑아내면 나이/성별 정보가 부족한 행을 보완하는 새로운 특징(feature)이 됩니다.

| 함수 | 설명 |
|---|---|
| Series.map(func) | 원소마다 함수 적용 |
| DataFrame.apply(func, axis=) | 행/열 단위로 함수 적용 |
| str.split(구분자) | 문자열 나누기 |
| str.strip() | 앞뒤 공백 제거 |

## 9. 데이터 재구조화 (Pivot / Melt / Stack / Unstack)

### 📖 이론 설명
`stack`/`unstack`은 **인덱스와 컬럼 사이를 서로 이동**시킵니다(unstack: 인덱스→컬럼, stack: 컬럼→인덱스). `pivot`은 `pivot(index=, columns=, values=)`으로 표를 다시 짜지만, **(index, columns) 조합에 중복 값이 있으면 에러**가 납니다. 이럴 때는 집계 함수를 같이 지정할 수 있는 `pivot_table(aggfunc=...)`을 씁니다. `melt`는 반대로 여러 컬럼을 하나의 '변수/값' 컬럼 쌍으로 길게 늘어뜨립니다(wide → long).

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| df.unstack(level=) | 인덱스를 컬럼으로 이동 |
| df.stack(level=) | 컬럼을 인덱스로 이동 |
| df.pivot(index=,columns=,values=) | 중복 없을 때만 사용 가능 |
| df.pivot_table(...,aggfunc=) | 중복 있어도 집계해서 사용 가능 |
| df.melt(id_vars=,var_name=,value_name=) | wide → long 변환 |

### 💻 예제 코드

````
data = {'날짜': ['2023-01-01']*2 + ['2023-01-02']*2,
        '지역': ['서울', '부산', '서울', '대구'],
        '판매량': [100, 50, 120, 40]}
df = pd.DataFrame(data)

pivot_df = df.pivot_table(index='날짜', columns='지역', values='판매량', aggfunc='sum')
print(pivot_df)

melted = pivot_df.reset_index().melt(id_vars='날짜', var_name='지역', value_name='판매량')
print(melted)
````

## 10. 데이터 병합 (Merge & Concat)

### 📖 이론 설명
`concat`은 단순히 이어붙이기(행 방향 `axis=0` 또는 열 방향 `axis=1`)이고, `merge`는 **공통 키(key) 값을 기준으로 관계형 데이터베이스처럼 조인**합니다. `how='inner'`는 양쪽에 다 있는 키만, `'outer'`는 전체, `'left'`/`'right'`는 한쪽 기준으로 결과를 만듭니다. 컬럼명이 다르면 `left_on`/`right_on`을, 인덱스를 기준으로 합칠 땐 `left_index=True`/`right_index=True`를 씁니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| pd.concat([a,b], axis=0/1) | 단순 이어붙이기 |
| pd.concat(..., join='inner'/'outer') | 공통 컬럼만 / 전체 컬럼 |
| pd.merge(a,b,on=,how=) | 키 기준 조인 |
| left_on/right_on | 컬럼명이 다를 때 |
| left_index/right_index | 인덱스 기준 조인 |

### 💻 예제 코드

````
test = pd.read_csv('../data/test.csv')  # PassengerId, LuggageCount, HasPet

df_inner = pd.merge(train, test, on='PassengerId', how='inner')
print(df_inner.shape)  # test.csv에 있는 60명만 남음

df_left = pd.merge(train, test, on='PassengerId', how='left')
print(df_left.shape)   # train 전체가 유지되고 없는 값은 NaN
````

## 13. 시계열 데이터 (Time Series)

### 📖 이론 설명
날짜 문자열은 `pd.to_datetime()`으로 실제 날짜 타입으로 바꿔야 `.dt.year`, `.dt.month` 같은 접근자와 `resample`(기간 단위 재집계), `rolling`(이동평균) 같은 시계열 전용 기능을 쓸 수 있습니다. 시계열 데이터의 결측치는 단순 평균 대체보다 **앞/뒤 값으로 채우거나(`bfill`/`ffill`) 선형 보간(`interpolate`)** 하는 경우가 많습니다 - 시간 순서에 의미가 있기 때문입니다.

> 💡 **왜 시험에 나오는가**: AICE Associate의 핵심 50개 예제는 Titanic류의 정형 데이터가 중심이라 시계열 전용 문제의 비중은 크지 않지만, 날짜 컬럼이 포함된 데이터가 나오면 `pd.to_datetime()`으로 변환 후 `.dt.year`/`.dt.month`를 새 피처로 추가하라는 식의 응용 문항이 나올 수 있습니다. 알아두면 손해 볼 것 없는 보완 지식으로 접근하면 됩니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| pd.to_datetime(col) | 문자열 → datetime 변환 |
| .dt.year/.dt.month | 날짜 속성 추출 |
| df.resample('M').sum() | 월 단위 재집계 |
| .rolling(window=n).mean() | n기간 이동평균 |
| .interpolate() | 선형 보간 |
| .bfill()/.ffill() | 뒤/앞 값으로 채우기 |

### 💻 예제 코드

````
dates = pd.date_range('2023-01-01', periods=6, freq='D')
ts = pd.Series([10, np.nan, 12, np.nan, 15, 16], index=dates)
print(ts.interpolate())
print(ts.ffill())
````

## 14. 데이터 삭제 & 인덱싱

### 📖 이론 설명
`drop_duplicates()`는 완전히 똑같은 행(또는 지정한 컬럼 기준 중복)을 제거합니다. `set_index`로 특정 컬럼을 인덱스로 승격시키면 `loc`로 그 값을 바로 조회할 수 있고, 여러 컬럼을 인덱스로 지정하면 MultiIndex가 됩니다. `reset_index(drop=True)`는 반대로 인덱스를 버리고 0부터 다시 매깁니다 - `train_test_split` 뒤 인덱스가 뒤섞였을 때 자주 씁니다.

> 💡 **왜 시험에 나오는가**: `drop_duplicates()`는 EDA 초반 '중복 데이터 제거' 문항에 직접 나오고, `reset_index(drop=True)`는 데이터를 나누거나(`train_test_split`) 필터링한 뒤 이어지는 코드가 인덱스 어긋남으로 에러 나는 것을 막아주는 실무 필수 습관이라 뒤쪽 머신러닝 노트북(04~08번)에서도 계속 재사용됩니다.

### 🧩 핵심 개념 정리
| 함수 | 설명 |
|---|---|
| df.drop_duplicates(subset=) | 중복 행 제거 |
| df.set_index('col') | 컬럼을 인덱스로 |
| df.reset_index(drop=True) | 인덱스를 0부터 재부여 |

### 💻 예제 코드

````
dup_df = pd.DataFrame({'a': [1, 1, 2], 'b': [3, 3, 4]})
print(dup_df.drop_duplicates())

named = train.set_index('PassengerId')
print(named.loc[1])  # PassengerId==1인 행 바로 조회
````

## 15. 실전 심화 테크닉 (pandas_exercises 대조 보완)

### 📖 이론 설명
지금까지 배운 함수들만으로도 대부분의 문제를 풀 수 있지만, 실무에서 아주 자주 마주치는 몇 가지 패턴을 더 소개합니다. 컬럼마다 **다른 집계 함수**를 적용하고 싶을 때, 그룹을 **하나씩 순회하며 확인**하고 싶을 때, **csv를 잘못 저장/로딩해서 생기는 'Unnamed' 컬럼** 문제, **헤더가 없는 파일**을 읽어야 할 때 등은 이론 수업에서는 잘 안 다루지만 실제 데이터를 만지다 보면 거의 매번 마주치는 상황들입니다.

### 🧩 핵심 개념 정리
| 상황 | 함수/문법 |
|---|---|
| 컬럼마다 다른 집계함수 적용 | `groupby(...).agg({'col1': 'mean', 'col2': 'max'})` |
| 그룹을 하나씩 순회 | `for name, group in df.groupby(...):` |
| 그룹별 비율(%) 계산 | `groupby(...)['col'].value_counts(normalize=True)` |
| 컬럼 순서 재배치 | `df[['원하는', '순서', '리스트']]` |
| 'Unnamed' 컬럼 제거 | `df.loc[:, ~df.columns.str.contains('^Unnamed')]` |
| 인덱스 범위 확장/재정렬 | `df.reindex(range(n))` |
| 헤더 없는 파일 읽기 | `pd.read_csv(path, header=None, names=[...])` |
| 인덱스 중복 확인/제거 | `df.index.duplicated()` |

### 💻 예제 코드

````
# 컬럼마다 다른 집계함수 적용
print(train.groupby('Pclass').agg({'Age': 'mean', 'Fare': 'max'}))

# 그룹을 하나씩 순회하기
for name, group in train.groupby('Pclass'):
    print(f'Pclass={name}: {len(group)}명, 평균 나이={group["Age"].mean():.1f}')

# 그룹별 비율 계산 (생존 여부를 Pclass별 비율로)
print(train.groupby('Pclass')['Survived'].value_counts(normalize=True))

# 'Unnamed' 컬럼 제거 (index=True로 저장했다가 다시 읽으면 흔히 생김)
messy = pd.DataFrame({'a': [1, 2], 'Unnamed: 0': [0, 1]})
messy = messy.loc[:, ~messy.columns.str.contains('^Unnamed')]
print(messy)
````
